### Part 1 (1 point)

Load the Helsinki temperature data from the file `data/helsinki-vantaa.csv`.

- Read the data into a variable called `data` using pandas
- Parse the dates from the column `'DATE'` and set the dates as the index in the DataFrame 

```python
# Import pandas library
import pandas as pd

# Load the Helsinki temperature data from file
file_path = 'data/helsinki-vantaa.csv'
data = pd.read_csv(file_path)

# Parse the dates and set the index
data['DATE'] = pd.to_datetime(data['DATE'], format='%Y-%m-%d')
data = data.set_index('DATE')
```

```python
# The lines in graph are not how they are supposed to be but the round markers seems to be rigth

import pandas as pd
import matplotlib.pyplot as plt

# Read the data into a pandas DataFrame
file_path = pd.read_csv('data/helsinki-vantaa.csv')

# Convert the DATE column to a pandas datetime object and set it as the index
file_path['DATE'] = pd.to_datetime(file_path['DATE'])
file_path.set_index('DATE', inplace=True)

# Filter the DataFrame to only include data from 1988 onwards
file_path = file_path[file_path.index >= '1988']

# Create a line plot of the monthly mean temperatures
fig, ax = plt.subplots(figsize=(14, 6))
ax.plot(file_path['TEMP_C'], linestyle='-', marker='o', color='black')
ax.set_title('Helsinki-Vantaa Airport')
ax.set_xlabel('Time')
ax.set_ylabel('Temperature (Celsius)')

# Add a background grid
ax.grid(True)

# Save the figure as a PNG file
outputfp = 'temp_line_plot.png'
plt.savefig(outputfp)
```
